```python
import timeit
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn import datasets, linear_model
from math import tanh, log

X, y = datasets.load_diabetes(return_X_y=True)
raw= X[:,None, 2]
```

```python
max_raw = max(raw)
min_raw = min(raw)
scaled = (2*raw-max_raw-min_raw)/(max_raw-min_raw)
```

```python
fig, axs = plt.subplots(2,1, sharex=True)
axs[0].hist(raw)
axs[1].hist(scaled)
```

```python
def train_raw():
    linear_model.LinearRegression().fit(raw, y)

def train_scaled_raw():
    linear_model.LinearRegression().fit(scaled, y)
```

```python
raw_time = timeit.timeit(train_raw, number=100)
scaled_raw_time = timeit.timeit(train_scaled_raw, number=100)

print(f"Raw data took {raw_time} seconds to train")
print(f"Scaled raw data took {scaled_raw_time} seconds to train")
```

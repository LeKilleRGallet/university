```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler

iris = sns.load_dataset('iris')

scaler=StandardScaler()
scaler=scaler.fit_transform(iris.drop('species',axis=1))

cov_mat=np.cov(scaler.T)
cov_mat
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot(x=iris['petal_length'],y=iris['petal_width'])
sns.jointplot(x=scaler[:,2],y=scaler[:,3])
```

```python
eigen_vals,eigen_vecs=np.linalg.eig(cov_mat)
```

```python
eigen_vals
```

```python
eigen_vecs
```

```python
var_expl=[]
for i in eigen_vals:
    var_expl.append(((i/sum(eigen_vals))*100))
```

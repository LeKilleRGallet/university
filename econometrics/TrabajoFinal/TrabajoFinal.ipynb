```r
noremu <- c('QUEHACERES_TIEMPO','AUTOCONSTRUCCION_TIEMPO','ROPA_TIEMPO','ANCIANOS_TIEMPO','NINYOS_TIEMPO','QUEHACERES_FUERA_TIEMPO','ANIMALES_TIEMPO')

df[noremu][is.na(df[noremu])] <- 0

df[c('EXTRAS_TIEMPO','HORAS_SEG_TRABAJO')][is.na(df[c('EXTRAS_TIEMPO','HORAS_SEG_TRABAJO')])] <- 0

df <- df[!is.na(df$INGRESOS),]

row.names(df) <- NULL
```

```r
library(tidyverse)
```

```r
df <- df %>% mutate(HORAS_REMUNERADAS = HORAS_SEG_TRABAJO+HORAS_PRIM_TRABAJO+EXTRAS_TIEMPO)

df <- df %>% mutate(HORAS_NO_REMUNERADAS = QUEHACERES_TIEMPO+AUTOCONSTRUCCION_TIEMPO+ROPA_TIEMPO+ANCIANOS_TIEMPO+NINYOS_TIEMPO+QUEHACERES_FUERA_TIEMPO+ANIMALES_TIEMPO)

df <- df %>% mutate(EXPERIENCIA = EDAD - EDUCACION - 6 )
```

```r
df[noremu]<- NULL

df[c('EXTRAS_TIEMPO','HORAS_SEG_TRABAJO','HORAS_PRIM_TRABAJO')]<- NULL
```

```r
df
```

```r
df$ANALFABETISMO <- ifelse(df$ANALFABETISMO == 2, 1, 0)
df$GENERO <- ifelse(df$GENERO == 2, 1, 0)
df$SINDICADO <- ifelse(df$SINDICADO == 1, 1, 0)
df$SEGUNDO_TRABAJO <- ifelse(df$SEGUNDO_TRABAJO == 1, 1, 0)
```

```r
sum(df$INGRESOS == 0)
```

```r
df <- df[!df$INGRESOS == 0,]
df <- df[!df$HORAS_REMUNERADAS == 0,]
```

```r
sum(df$INGRESOS == 0)
```

```r
df[c('POBREZA','CAMBIO_TRABAJO','CONTRATO','TERMINO_CONTRATO','MODALIDAD_CONTRATO')]<- NULL
df$ln_INGRESOS <- log(df$INGRESOS)
df$TIEMPO_EMPRESA <- df$TIEMPO_EMPRESA/12

df$ING_H <- (df$INGRESOS/4)/df$HORAS_REMUNERADAS

df$ln_ING_H <- log(df$ING_H)

df['INGRESOS'] <- NULL
```

```r
df$INDIGENA <- ifelse(df$ETNIA == 1, 1, 0)
# df$GITANO <- ifelse(df$ETNIA == 2, 1, 0)
# df$RAIZAL <- ifelse(df$ETNIA == 3, 1, 0)
# df$PALENQUERO <- ifelse(df$ETNIA == 4, 1, 0)
df$AFRO <- ifelse( df$ETNIA == 3 | df$ETNIA == 4 | df$ETNIA == 5, 1, 0)
```

```r
descritive <- rbind(summary(df$ln_ING_H), summary(df$EDUCACION), summary(df$EXPERIENCIA), summary(df$EXPERIENCIA2), summary(df$TIEMPO_EMPRESA), summary(df$HORAS_REMUNERADAS), summary(df$HORAS_NO_REMUNERADAS), summary(df$INDIGENA), summary(df$AFRO), summary(df$ANALFABETISMO), summary(df$GENERO), summary(df$SINDICADO), summary(df$SEGUNDO_TRABAJO))
unlist(lapply(descritive, is.numeric))

descritive <- cbind(descritive, lapply(df[,unlist(lapply(df, is.numeric))],var), lapply(df[,unlist(lapply(df, is.numeric))],sd))
colnames(descritive) <- c("Min.","1st Qu.", "Median","Mean","3rd Qu.","Max.","Var","SD")
rownames(descritive) <- c('Salario','Educacion','Experiencia','Experiencia2','Tiempo Empresa','Horas Remuneradas','Horas No Remuneradas','Indigena','Afro','Analfabetismo','Genero','Sindicado','Segundo Trabajo')
```

```r
colnames(df)[colnames(df) == "P3246"] <- "POBREZA"
colnames(df)[colnames(df) == "P6160"] <- "ANALFABETISMO"
colnames(df)[colnames(df) == "P6080"] <- "ETNIA"
colnames(df)[colnames(df) == "ESC"] <- "EDUCACION"
colnames(df)[colnames(df) == "P6020"] <- "GENERO"
colnames(df)[colnames(df) == "P7480S3A1"] <- "QUEHACERES_TIEMPO"
colnames(df)[colnames(df) == "P7480S9A1"] <- "AUTOCONSTRUCCION_TIEMPO"
colnames(df)[colnames(df) == "P7480S7A1"] <- "ROPA_TIEMPO"
colnames(df)[colnames(df) == "P7480S6A1"] <- "ANCIANOS_TIEMPO"
colnames(df)[colnames(df) == "P7480S5A1"] <- "NINYOS_TIEMPO"
colnames(df)[colnames(df) == "P7480S4A1"] <- "QUEHACERES_FUERA_TIEMPO"
colnames(df)[colnames(df) == "P7480S1A1"] <- "ANIMALES_TIEMPO"
colnames(df)[colnames(df) == "INGLABO"] <- "INGRESOS"
colnames(df)[colnames(df) == "P6640S1"] <- "EXTRAS_TIEMPO"
colnames(df)[colnames(df) == "P7180"] <- "SINDICADO"
colnames(df)[colnames(df) == "P7130"] <- "CAMBIO_TRABAJO"
colnames(df)[colnames(df) == "P7045"] <- "HORAS_SEG_TRABAJO"
colnames(df)[colnames(df) == "P7040"] <- "SEGUNDO_TRABAJO"
colnames(df)[colnames(df) == "P6850"] <- "HORAS_PRIM_TRABAJO"
colnames(df)[colnames(df) == "P6440"] <- "CONTRATO"
colnames(df)[colnames(df) == "P6460"] <- "TERMINO_CONTRATO"
colnames(df)[colnames(df) == "P6450"] <- "MODALIDAD_CONTRATO"
colnames(df)[colnames(df) == "P6426"] <- "TIEMPO_EMPRESA"
```

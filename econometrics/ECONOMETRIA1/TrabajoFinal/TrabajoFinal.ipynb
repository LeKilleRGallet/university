```r
#create a empty df named montecarlo_df
montecarlo_df <- data.frame(matrix(ncol = 12, nrow = 0))
#create 12 empty columns in montecarlo_df

colnames(montecarlo_df) <- colnames(exogenous_df)

montecarlo_size = 10000

for (i in 1:montecarlo_size) {
    for (j in 1:12) {
        montecarlo_df[i,j] <- sample(exogenous_df[,j], size = 1)
    }
}

montecarlo_df
```

```r
ln_ing_h_predict <-  predict(fit, newdata = montecarlo_df)
```

```r
png("montecarlo.png")
hist(ln_ing_h_predict, main = "Montecarlo simulation", xlab = "Ln(income)", ylab = "Frequency", breaks = montecarlo_size/100)
abline(v = mean(ln_ing_h_predict), col = "red")
dev.off()
```

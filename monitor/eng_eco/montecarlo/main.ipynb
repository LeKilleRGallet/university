```python
import random
import matplotlib.pyplot as plt
```

```python
def npv(cash_flow, rate):
    """
    Calculate the net present value of a cash flow given a discount rate.
    """
    return sum(cash_flow[i] / (1 + rate) ** i for i in range(len(cash_flow)))
```

```python
NPV=[]

n=1000

for _ in  range(n):
    cashflow=[-random.randint(10,100), 10, 50, 30, 70,200, 50, 100]
    rate =  0.07 #random.uniform(0.001, 0.3)

    inv_in = -random.randint(10,300)

    cashflow.append(inv_in)

    # for i in range(10):
        
    #     income=random.randint(70,200)
    #     outcome=random.randint(60,150)
        
    #     cashflow.append(income-outcome)
    
    # print(cashflow)
    
    NPV.append(npv(cashflow, rate))
    
mean=sum(NPV)/len(NPV)

print("The mean of the NPV is: ", mean)

plt.hist(NPV, bins=100)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
```

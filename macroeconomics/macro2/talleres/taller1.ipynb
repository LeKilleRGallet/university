```python
df=pd.read_table('IIP.txt', sep='\t')
df
```

```python
plt.plot(df['YEAR'], df['IIP'])
plt.xlabel('Year')
plt.ylabel('IIP')
plt.title('IIP by Year in Colombia (M$USD)')
plt.annotate('source: banrep' , xy=(0.5, 0.05), xycoords='axes fraction', fontsize=10, ha='center')
plt.show()
```

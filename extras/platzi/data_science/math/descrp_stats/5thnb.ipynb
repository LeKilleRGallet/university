```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
import pandas as pd

iris=sns.load_dataset('iris')
```

```python
sns.pairplot(iris,hue='species')
```

```python
scaler=StandardScaler()
scaler=scaler.fit_transform(iris.drop('species',axis=1))
scaler.T
```

```python
cov_matrix=np.cov(scaler.T)
sns.heatmap(cov_matrix,annot=True,fmt='.2f')
```

```python
scaler=pd.DataFrame(scaler,columns=iris.columns[:-1])
sns.heatmap(scaler.cov(),annot=True,fmt='.2f')
```

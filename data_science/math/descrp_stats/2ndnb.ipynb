```python
import pandas as pd
import seaborn as sns

iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris, hue='species')
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris, hue='species')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris, hue='species')
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris)
```

```python
sns.barplot(x='species',y='sepal_length',data=iris)
```

```python
sns.barplot(x='species',y='sepal_width',data=iris)
```

```python
sns.barplot(x='species',y='petal_length',data=iris)
```

```python
#min-max nomalization
iris_norm=iris.copy()
iris_norm['sepal_length']=(iris_norm['sepal_length']-iris_norm['sepal_length'].min())/(iris_norm['sepal_length'].max()-iris_norm['sepal_length'].min())
iris_norm['sepal_width']=(iris_norm['sepal_width']-iris_norm['sepal_width'].min())/(iris_norm['sepal_width'].max()-iris_norm['sepal_width'].min())
iris_norm['petal_length']=(iris_norm['petal_length']-iris_norm['petal_length'].min())/(iris_norm['petal_length'].max()-iris_norm['petal_length'].min())
iris_norm['petal_width']=(iris_norm['petal_width']-iris_norm['petal_width'].min())/(iris_norm['petal_width'].max()-iris_norm['petal_width'].min())
iris_norm.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris_norm, hue='species')
```

```python
sns.boxplot(x='species',y='sepal_length',data=iris_norm)
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris_norm, hue='species')
```

```python
sns.barplot(x='species',y='sepal_length',data=iris_norm)
```

```python
#z-score normalization
iris_zscore=iris.copy()
iris_zscore['sepal_length']=(iris_zscore['sepal_length']-iris_zscore['sepal_length'].mean())/iris_zscore['sepal_length'].std()
iris_zscore['sepal_width']=(iris_zscore['sepal_width']-iris_zscore['sepal_width'].mean())/iris_zscore['sepal_width'].std()
iris_zscore['petal_length']=(iris_zscore['petal_length']-iris_zscore['petal_length'].mean())/iris_zscore['petal_length'].std()
iris_zscore['petal_width']=(iris_zscore['petal_width']-iris_zscore['petal_width'].mean())/iris_zscore['petal_width'].std()
iris_zscore.head()
```

```python
sns.scatterplot(x='sepal_length',y='petal_length',data=iris_zscore, hue='species')
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=iris_zscore, hue='species')
```

```python
sns.barplot(x='species',y='sepal_length',data=iris_zscore)
```
